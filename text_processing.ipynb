+ (D) STEMMING / LEMMATIZATION 

+ STEMMING 

```python
from nltk.stem import PorterStemmer

# Step 1: Initialize a list of sample words that we want to stem (truncate)
words = ["runner", "burnt", "happily", "easily", "lovely", "flying", "leaf"]

# Step 2: Create an instance of the PorterStemmer object
# PorterStemmer is a fast, rule-based algorithm that strips common suffixes like 'ing', 'ly', 'ed'
stemmer = PorterStemmer()

# Step 3: Use list comprehension to iterate through each word in the list.
# The loop passes each 'word' to stemmer.stem(word) to chop off its ending 
# and automatically saves the processed root forms into the 'stemmed_words' list.
stemmed_words = [stemmer.stem(word) for word in words]

# Step 4: Print the final list of stemmed tokens to the console
print(stemmed_words)
```

+ LEMMATIZATION 

```python
import spacy

# Step 1: Load the pre-trained small English language model package
# This package contains the grammatical rules and dictionaries required for lemmatization
nlp = spacy.load("en_core_web_sm")

# Step 2: Define the raw text string that we want to analyze
text = " I am running my best time"

# Step 3: Pass the text through the 'nlp' pipeline object
# This automatically parses the text, tokenizes it, and calculates its linguistic features
doc = nlp(text)

# Step 4: Use list comprehension to loop through every single token in the 'doc' object.
# The '.lemma_' attribute converts each word back into its base/dictionary form.
# All the processed base words are automatically saved inside the 'lemmatized_words' list.
lemmatized_words = [token.lemma_ for token in doc]

# Step 5: Print the final list of lemmatized words to the console
print(lemmatized_words)
```

```python
# Assuming 'doc' is already created using a spaCy model (e.g., doc = nlp("Your text here"))

# Step 1: Loop through every single word/punctuation (token) found in the processed 'doc' object.
for token in doc:
    
    # Step 2: Print the token's text, its Part-of-Speech tag, and its lemmatized root form.
    # The :<15 and :<10 syntax applies left-alignment spacing to create neat, clean columns.
    # - token.text:<15  -> Prints the raw word and pads it with spaces to take up exactly 15 characters.
    # - token.pos_:<10  -> Prints the grammatical category (Noun, Verb, etc.) padded to exactly 10 characters.
    # - token.lemma_    -> Prints the baseline dictionary/root version of the word at the end.
    print(f"{token.text:<15}{token.pos_:<10}{token.lemma_}")
```

+ 5th REMOVE PUNCTUATION 

```python
import string

# Step 1: Ek sample text string banaya jisme comma (,) aur full stop (.) jaise punctuations hain
text = """ I have a news , and it is a great . I am going to share it , but only with people who have helped me .
And especially these , who have helped me stay at the top"""

# Step 2: str.maketrans() ka use karke ek 'Translation Table' (mapping dictionary) banayi.
# iske shuruat ke 2 parameters khali ('', '') hain, aur teesre parameter me humne string.punctuation diya hai.
# Iska matlab: "string.punctuation me jitne bhi characters hain (जैसे !, ., ,, ?), unhe 'None' (delete) me map kar do."
translator = str.maketrans("", "", string.punctuation)

# Step 3: .translate() function ka use karke us mapping table ko apne 'text' par apply kiya.
# Yeh text ke andar jayega aur har ek punctuation mark ko dhoodh kar use string se permanent hata (strip) dega.
processed_text = text.translate(translator)

# Step 4: Punctuation-free clean text ko screen par print kiya
print(len(text))
print(processed_text)
print(len(processed_text))
```

+ (6) N-GRAMS 

```python
from nltk.tokenize import word_tokenize
from nltk.util import ngrams

# Step 1: Define a sample text string to be broken down into N-grams
sample_text = " My name is xyz , i have a 10sec record in running 100 meters"

# Step 2: Convert the raw sentence into a list of individual words and punctuation marks (tokens)
tokens = word_tokenize(sample_text)

# Step 3: Generate 10-grams (groups of 10 consecutive words) using the ngrams() function.
# The number 10 means the sliding window will capture exactly 10 tokens at a time.
# We wrap it in list() because ngrams() returns a generator object that needs conversion to be viewable.
tengrams = list(ngrams(tokens, 10))

# Step 4: Print the final list containing the generated 10-gram tuples to the console
print(tengrams)
```

+ (7) POS Tagging 

```python
import nltk
from nltk import pos_tag

# Step 1: Download the specific trained model required for Part-of-Speech (POS) tagging.
# 'averaged_perceptron_tagger_eng' contains the pre-trained weights and data mapping 
# that allows NLTK to read English sentences and identify nouns, verbs, adjectives, etc.
# Note: This download only needs to be run once in your environment.
nltk.download("averaged_perceptron_tagger_eng")
```

```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag

# Step 1: Ek sample text string banaya jiske har ek word ka grammar (POS) pehchanna hai
text = "I am running my best time"

# Step 2: word_tokenize() ka use karke text ko alag-alag tokens (words) me toda.
# Output: ['I', 'am', 'running', 'my', 'best', 'time']
tokens = word_tokenize(text)

# Step 3: pos_tag() function me tokens ki list pass ki.
# Yeh function har ek word ko uske grammatical role (Noun, Verb, Adjective) ke hisab se ek shortcut tag de deta hai.
tagged = pos_tag(tokens)

# Step 4: Final output ko print kiya, jo tuples ki ek list hoti hai -> (Word, Tag)
print(tagged)
```
